```python
#  Import library, this example for Decision tree classifires
#  DecisionTreeClassifier is a Supervised learing algoritm

import pandas as pd

df = pd.read_csv('E:/ML/New/companydata.csv')

df


# find the based on company, jobtitle and degree which company is providing salary more than 100$
```

```python
#  get information about dataset

df.info()
```

```python
#  describe the dataset
df.describe()
```

```python
#  check if any null values
df.isnull().sum()
```

```python
#  convert the salarymorethan100usd to yes=1 and no=0


df.loc[df["salarymorethan100USD"]=='yes', "salarymorethan100USD"]=1
df.loc[df["salarymorethan100USD"]=='no', "salarymorethan100USD"]=0
df
```

```python
#  define the feature and traget variable that is 

#   x=features var
#   y=target var

x = df.drop('salarymorethan100USD', axis='columns')
y = df['salarymorethan100USD']
```

```python
#x
```

```python
# y varaiable type is flot need to convert into Int as below

y=y.astype('int')
```

```python
y
```

```python
#  Label encoder used for convert the text data into number


from sklearn.preprocessing import LabelEncoder
```

```python
# 

LE_Company = LabelEncoder()
LE_jobtital = LabelEncoder()
LE_degree = LabelEncoder()
```

```python
#  it will conver the default values to all as below

x ['company_dummy']=LE_Company.fit_transform(x['Company'])
x ['jobtital_dummy']=LE_Company.fit_transform(x['jobtital'])
x ['degree_dummy']=LE_Company.fit_transform(x['degree'])

x.head(9)
```

```python
#  drop the old columns and get new freatures var


x_new = x.drop(['Company','jobtital','degree'], axis='columns')

x_new
```

```python
#   sklearn.model_selection used for train and test the data

#from sklearn.model_selection import train_test_split
# x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,random_state=10)
```

```python
# import decision tree classifiler algorithm, its supervised learing algorithm

from sklearn import tree

model = tree.DecisionTreeClassifier()
```

```python
#  fit the model that is train your model

model.fit(x_new, y)
```

```python
#  check the score

model.score(x_new,y)
```

```python
#  check the any values for salary is more than 100usd
#  provide the 3 parm

model.predict([[2,1,0]])
```
