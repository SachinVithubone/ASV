```python
# 1. add lib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 2. Label and amount
Exp_amount=[7500, 2500, 480, 700, 1000]
Exp_label=['homeRent', 'Food', 'EBill', 'Car', 'Eduction']

plt.pie(Exp_amount, labels=Exp_label)
```

```python
# 3. Draw pie chart as perfect circle

plt.pie(Exp_amount,labels=Exp_label, shadow=True)
plt.axis("equal")
plt.show()
```

```python
# 4. Show the precentage for every pie Specify radius to increase chart size

plt.axis("equal")
plt.pie(Exp_amount, labels=Exp_label,shadow=True, autopct='%1.1f%%',radius=1.5)
plt.show()
```

```python
# 5. Explode

plt.axis("equal")
plt.pie(Exp_amount,labels=Exp_label, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2])
plt.show()
```

```python
# 6. counterclock and angle properties


plt.axis("equal")
plt.pie(Exp_amount,labels=Exp_label, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],
        counterclock=True, startangle=45)
plt.show()
```
