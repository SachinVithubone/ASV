```python
# Replace the single value as below
Wdf = Wdf.replace(-99999, value=np.NaN)
Wdf
```

```python
# Now replace the all NaN to Zero '0'

Wdf = Wdf.replace(to_replace=[-99999, -88888], value=0)

Wdf
```

```python
# Replace NaN value to 0 for belwo cloun

Wdf.temperature = Wdf.temperature.fillna(0)
Wdf.windspeed = Wdf.windspeed.fillna(0)

Wdf
```

```python
# when windspeed is 6 mph, 7 mph etc. & temperature is 32 F, 28 F etc.
Wdf = Wdf.replace({'temperature': '[A-Za-z]', 'windspeed': '[a-z]'},'', regex=True) 

Wdf
```
